```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import csv file
iris = pd.read_csv("Iris.csv")
```

```python
#check first 5 rows
iris.head()
```

```python
#check numerics
iris.describe()
```

```python
#basic scatter plot with Pandas
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm", title="Sepal Scatter Plot")
```

```python
#add color and legend using Seaborn
sns.FacetGrid(iris, hue="Species", height=6).map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend().set(title='Plot with Color')
```

```python
#boxplots of PetalLength(Width) vs Species
fig, axes = plt.subplots(ncols=2, figsize=(10, 5))
sns.boxplot(ax=axes[0], x="Species", y="PetalLengthCm", data=iris).set(title="PetalLength per Species Boxplot")
sns.boxplot(ax=axes[1], x="Species", y="PetalWidthCm", data=iris).set(title="PetalWidth per Species Boxplot")
```

```python
#To see the actual distribution of the individual data, use Seaborn's striplot

fig, axes = plt.subplots(ncols=2, figsize=(10,5))
sns.boxplot(ax=axes[0],x="Species", y="PetalLengthCm", data=iris).set(title="PetalLength Distribution")
sns.stripplot(ax=axes[0],x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
sns.boxplot(ax=axes[1],x="Species", y="PetalWidthCm", data=iris).set(title="PetalWidth Distribution")
sns.stripplot(ax=axes[1],x="Species", y="PetalWidthCm", data=iris, jitter=True, edgecolor="gray")
```
